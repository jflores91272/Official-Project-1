```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv(r'/Users/jonathanflores/Documents/Flatiron/Gitted_project/zippedData/tn.movie_budgets.csv')
df
```

```python
df['production_budget'] =df['production_budget'].apply(lambda x: x.replace('$',' '))
df['domestic_gross'] =df['domestic_gross'].apply(lambda x: x.replace('$',' '))
df['worldwide_gross'] =df['worldwide_gross'].apply(lambda x: x.replace('$',' '))
df['production_budget'] =df['production_budget'].apply(lambda x: x.replace(',',' '))
df['domestic_gross'] =df['domestic_gross'].apply(lambda x: x.replace(',',' '))
df['worldwide_gross'] =df['worldwide_gross'].apply(lambda x: x.replace(',',' '))
df
```

```python
df['budget']= df['production_budget'].str.replace(' ', '')
df
```

```python
df['Domesticgross']=df['domestic_gross'].str.replace(' ', '')
df['WorldWidegross']=df['worldwide_gross'].str.replace(' ', '')
df
```

```python
df2 = df.drop(['production_budget', 'domestic_gross'], axis=1)
```

```python
df2
```

```python
df2["budget"] = pd.to_numeric(df["budget"], downcast="float")
```

```python
df2.dtypes
```

```python
df2["WorldWidegross"] = pd.to_numeric(df["WorldWidegross"], downcast="float")
```

```python
df2.dtypes
```

```python
df2["ProfitMargin"] = df2["WorldWidegross"]/df2["budget"]
df2
```

```python
df2['ProfitMargin'] = df2['ProfitMargin'].apply(lambda x: x*100)
```

```python
df10=df2.sort_values(['ProfitMargin'], ascending=True)
df10.head(60)
```

```python
df3 = df2.sort_values(['WorldWidegross'], ascending=False)
df3.head(16)
```

```python
df4 = df3.head(16)
```

```python
df4.sort_values(['budget','WorldWidegross'], ascending=[False,False])
```

```python
ax = sns.scatterplot(x = 'movie', y = 'ProfitMargin',hue = 'budget', data = df4)
ax.tick_params(axis='x', labelrotation = 90)
sns.move_legend(ax,"upper left", bbox_to_anchor=(1, 1))
sns.color_palette("magma", as_cmap=True)

plt.show()
```

```python
ax=sns.barplot(data=df4, x="movie", y="ProfitMargin")
ax.tick_params(axis='x', labelrotation = 90)
```
